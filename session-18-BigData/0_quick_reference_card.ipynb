# Quick Reference Cards

### ACID Properties Quick Reference

| Property | One-liner |
|----------|-----------|
| Atomicity | All or nothing |
| Consistency | Rules always enforced |
| Isolation | Transactions don't interfere |
| Durability | Committed = Permanent |

### Storage Format Quick Reference

| Need | Use |
|------|-----|
| Fast inserts/updates | Row-based (CSV, RDBMS) |
| Fast analytics | Column-based (Parquet, ORC) |
